```python
'''1.What is polymorphism in Python? Explain how it is related to object-oriented programming.'''

# Ans
'''
Polymorphism is a fundamental concept in object-oriented programming (OOP) that allows objects of different classes to be 
treated as objects of a common superclass. In Python, polymorphism is closely related to OOP principles and is achieved 
through method overriding and method overloading.

Here's how polymorphism is related to object-oriented programming in Python:
Inheritance: Polymorphism is often realized through inheritance, which is one of the key principles of OOP. Inheritance allows
            you to create a new class (subclass or derived class) based on an existing class (base class or superclass). The
            derived class inherits the attributes and methods of the base class, making it possible to use polymorphism 
            effectively.

Method Overriding: In Python, polymorphism is primarily implemented through method overriding. Method overriding allows a 
            subclass to provide a specific implementation of a method that is already defined in its superclass. When you 
            call the overridden method on an object of the subclass, the subclass's implementation is executed, not the one
            from the superclass. This enables you to have different behavior for the same method in different classes, making
            code more flexible and extensible.
'''
# Code
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"
animals = [Dog(), Cat()]
for animal in animals:
    print(animal.speak())
```

```python
'''2.Describe the difference between compile-time polymorphism and runtime polymorphism in Python.'''

# Ans
'''
Compile-time polymorphism and runtime polymorphism are two different forms of polymorphism in Python and other programming 
languages. They refer to when the binding of a method (function) call to its implementation occurs. Here's the key difference 
between the two:

Compile-Time Polymorphism (Static Binding):
Compile-time polymorphism, also known as static binding or early binding, is determined at compile time.
In compile-time polymorphism, the decision about which method to call is made by the compiler based on the function name, 
the number, and types of arguments passed to it, and the function's signature (function overloading).
This type of polymorphism is typically associated with function or method overloading, where multiple functions or methods
with the same name exist in a class, but they have different parameter lists (different arguments).
The appropriate function or method to call is decided during compile time, and the specific function is bound to the call.
'''
# Code
class MathOperations:
    def add(self, a, b=0, c=0):
        return a + b + c
math_obj = MathOperations()
result1 = math_obj.add(2)        
result2 = math_obj.add(2, 3)       
result3 = math_obj.add(2, 3, 4)   
print(result1) 
print(result2)  
print(result3)  
```

```python
'''Runtime Polymorphism (Dynamic Binding):
Runtime polymorphism, also known as dynamic binding or late binding, is determined at runtime.
In runtime polymorphism, the decision about which method to call is deferred until the program is actually executed. 
This is typically achieved through method overriding, where a subclass provides a specific implementation for a method defined
in its superclass.
The appropriate method is selected at runtime based on the actual object's type, rather than the reference or variable's type.
This allows for more flexible and extensible code.'''

# Code
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

# Polymorphism in action
animals = [Dog(), Cat()]
for animal in animals:
    print(animal.speak())
```

```python
'''7.Discuss the use of abstract methods and classes in achieving polymorphism in Python. 
Provide an example using the `abc` module.'''

# Ans
'''
Abstract methods and classes in Python play a crucial role in achieving polymorphism and enforcing a common interface among 
related classes. They are implemented using the abc (Abstract Base Classes) module in Python. An abstract method is a method 
that is declared in an abstract class but does not have a concrete implementation. Subclasses must provide concrete 
implementations for these abstract methods, ensuring that certain methods are available across related classes, even though
their behavior may differ.

We define an abstract class Animal that inherits from ABC (Abstract Base Class). Inside this abstract class, we declare an 
abstract method speak() using the @abstractmethod decorator. This method is declared but not implemented in the Animal class.

We create three concrete subclasses Dog, Cat, and Bird, each of which inherits from the Animal abstract class. These 
subclasses provide concrete implementations for the speak() method.

We create instances of the Dog, Cat, and Bird classes.

We demonstrate polymorphism by creating a list of animals and then iterating through the list, calling the speak() method on 
each object. Despite the abstract method in the Animal class, the concrete implementations in the subclasses ensure that the 
appropriate speak() method is called for each object.
'''

# Code
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def speak(self):
        pass
class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

class Bird(Animal):
    def speak(self):
        return "Tweet!"
my_dog = Dog()
my_cat = Cat()
my_bird = Bird()
animal_list = [my_dog, my_cat, my_bird]

for animal in animal_list:
    print(animal.__class__.__name__ + " says: " + animal.speak())
```

```python
'''10.What is the role of the `@abstractmethod` decorator in achieving polymorphism in Python? Provide an example.'''

# Ans
'''

The @abstractmethod decorator is part of Python's Abstract Base Classes (ABCs) module, which allows you to define abstract 
methods in a base class. Abstract methods are methods that don't have an implementation in the base class but must be 
implemented in the derived (sub) classes. This concept is significant in achieving polymorphism in Python because it enforces 
a common interface that subclasses must adhere to, enabling polymorphism where objects of different classes can be treated as
instances of a common base class.


In this example, we define an abstract base class Shape with an abstract method area(). The @abstractmethod decorator marks
the area() method as abstract, indicating that it must be implemented in any concrete subclass of Shape. The concrete 
subclasses, Circle and Rectangle, both implement the area() method, providing their own specific logic to calculate the area.

The print_area() function demonstrates polymorphism. It takes an argument of type Shape, which can be either a Circle or a
Rectangle. Since both Circle and Rectangle inherit from Shape and implement the area() method, we can use polymorphism to call
print_area() with instances of these subclasses, and it correctly calculates and prints the areas of circles and rectangles.
'''

# Code
from abc import ABC, abstractmethod

# Define an abstract base class with abstract method(s)
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14159265359 * self.radius**2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
def print_area(shape):
    print(f"Area: {shape.area()}")
circle = Circle(5)
rectangle = Rectangle(4, 6)
print_area(circle)   
print_area(rectangle) 
```

```python
'''13.Explain the use of the `super()` function in Python polymorphism. How does it help call methods of parent classes?'''

# Ans
'''
The super() function in Python is used to call methods and access attributes from a parent or superclass within a derived or
subclass. It plays a crucial role in achieving polymorphism, allowing you to build on and extend the behavior of a parent 
class while maintaining a common interface. Here's how it helps call methods of parent classes in the context of polymorphism:

Method Overriding:
In polymorphism, you often create a base class (superclass) with certain methods that establish a common interface or behavior.
Subclasses (derived classes) then inherit from the base class and provide their own implementations of these methods. This 
process is called method overriding.If a subclass wants to extend or modify the behavior of a method from the parent class 
while still using the original behavior, super() comes in handy.

Calling Parent Class Methods:
When you override a method in a subclass, you can use super() to call the same method from the parent class. This is useful 
when you want to build on the existing behavior or add extra functionality.By using super(), you ensure that you're calling
the method defined in the parent class, maintaining the integrity of the common interface defined by the base class.
'''


# Code
class Shape:
    def area(self):
        return 0

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        # Call the area method from the parent class (Shape)
        parent_area = super().area()
        return 3.14159265359 * self.radius**2 + parent_area

circle = Circle(5)
print(circle.area())
```

```python
'''14.Create a Python class hierarchy for a banking system with various account types (e.g., savings, checking, credit card)
and demonstrate polymorphism by implementing a common `withdraw()` method.'''

# Code
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount.")
        elif amount > self.balance:
            print("Insufficient funds.")
        else:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")

class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance, min_balance):
        super().__init__(account_number, balance)
        self.min_balance = min_balance

    def withdraw(self, amount):
        if self.balance - amount < self.min_balance:
            print("Withdrawal not allowed. Minimum balance requirement not met.")
        else:
            super().withdraw(amount)

class CheckingAccount(BankAccount):
    def __init__(self, account_number, balance):
        super().__init__(account_number, balance)
        self.overdraft_limit = -500

    def withdraw(self, amount):
        if amount > self.balance + self.overdraft_limit:
            print("Withdrawal not allowed. Overdraft limit exceeded.")
        else:
            super().withdraw(amount)

class CreditCardAccount(BankAccount):
    def __init__(self, account_number, balance, credit_limit):
        super().__init__(account_number, balance)
        self.credit_limit = credit_limit

    def withdraw(self, amount):
        if amount > self.balance + self.credit_limit:
            print("Withdrawal not allowed. Credit limit exceeded.")
        else:
            super().withdraw(amount)
savings = SavingsAccount("SAV123", 1000, 100)
checking = CheckingAccount("CHK456", 1500)
credit_card = CreditCardAccount("CC789", 200, 1000)
savings.withdraw(200) 
checking.withdraw(200)  
credit_card.withdraw(200)  

savings.withdraw(800)  
checking.withdraw(1500)  
credit_card.withdraw(1200) 
```

```python
'''15.Describe the concept of operatoroverloading in Python and how it relates to polymorphism.
Provide examples using operators like `+` and `*`.'''

# Ans
'''**Operator overloading** in Python refers to the ability to define and customize the behavior of built-in operators (e.g.,
`+`, `-`, `*`, `/`, `==`, `!=`, etc.) for user-defined objects. It allows you to redefine what these operators do when applied
to instances of your custom classes. Operator overloading is closely related to polymorphism, as it enables different classes
to respond to operators in a way that adheres to the principles of polymorphism.

Here's how operator overloading relates to polymorphism, with examples using the `+` and `*` operators:

1. **Polymorphic Behavior**:
   - Operator overloading allows different classes to respond to operators in a polymorphic way. This means that the same 
   operator can have different meanings for different classes, as long as the classes provide an appropriate implementation.
   - For example, the `+` operator can be used to concatenate strings, add numbers, or perform custom operations, depending on
   the context.

2. **Common Interface**:
   - Operator overloading enforces a common interface for operators across different classes. This is essential for achieving
    polymorphism because it ensures that different objects can be used interchangeably with the same operators.
   - This common interface enables you to write generic code that can work with various types of objects, making your code 
   more flexible and adaptable.

Here are examples of operator overloading with the `+` and `*` operators:'''

# **Operator Overloading with `+` Operator:**

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        if isinstance(other, Vector):
            # Vector addition
            return Vector(self.x + other.x, self.y + other.y)
        elif isinstance(other, (int, float)):
            # Scalar addition
            return Vector(self.x + other, self.y + other)
        else:
            raise ValueError("Unsupported operand type")

    def __str__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2  # Vector addition
v4 = v1 + 5   # Scalar addition

print(v3)  # Output: Vector(4, 6)
print(v4)  # Output: Vector(6, 7)


'''In this example, the `+` operator is overloaded in the `Vector` class to perform both vector addition (adding two `Vector` 
instances) and scalar addition (adding a scalar to a `Vector`). This demonstrates polymorphism, as the `+` operator works 
differently based on the types of its operands.'''

# **Operator Overloading with `*` Operator:**

class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __mul__(self, other):
        if isinstance(other, ComplexNumber):
            # Complex number multiplication
            return ComplexNumber(
                self.real * other.real - self.imag * other.imag,
                self.real * other.imag + self.imag * other.real
            )
        elif isinstance(other, (int, float)):
            # Scalar multiplication
            return ComplexNumber(self.real * other, self.imag * other)
        else:
            raise ValueError("Unsupported operand type")

    def __str__(self):
        return f"{self.real} + {self.imag}i"

c1 = ComplexNumber(2, 3)
c2 = ComplexNumber(4, 5)
c3 = c1 * c2  # Complex number multiplication
c4 = c1 * 2    # Scalar multiplication

print(c3)  # Output: -7 + 22i
print(c4)  # Output: 4 + 6i

```

```python
'''16.What is dynamic polymorphism, and how is it achieved in Python?'''

# Ans
'''
Dynamic polymorphism, also known as runtime polymorphism or late binding, is a fundamental concept in object-oriented 
programming. It allows different classes to be treated as instances of a common base class, and method calls are resolved at 
runtime based on the actual type of the object. Dynamic polymorphism enables you to write code that can work with objects of
various classes, as long as they adhere to a common interface (i.e., share method names and signatures).

In Python, dynamic polymorphism is achieved through the use of inheritance and method overriding. Here's how it works:

Inheritance: You create a class hierarchy where subclasses inherit from a common base class. The base class defines a method 
that serves as a common interface.
Method Overriding: Subclasses override the methods defined in the base class. This means they provide their own 
implementations of these methods, which may have different behavior specific to each subclass.

Late Binding: When you call a method on an object, Python determines at runtime which method implementation to invoke based on 
the actual type of the object. This is late binding, and it's a key feature of dynamic polymorphism.
'''

# Code
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

# Polymorphic function
def make_animal_speak(animal):
    return animal.speak()

# Create instances of different animal types
dog = Dog()
cat = Cat()

# Call the polymorphic function
print(make_animal_speak(dog))  
print(make_animal_speak(cat))  
```

```python
'''17.Create a Python class hierarchy for employees in a company (e.g., manager, developer, designer) 
and implement polymorphism through a common `calculate_salary()` method.'''

# Code
class Employee:
    def __init__(self, name, role, experience):
        self.name = name
        self.role = role
        self.experience = experience

    def calculate_salary(self):
        raise NotImplementedError("Subclasses must implement this method")

class Manager(Employee):
    def calculate_salary(self):
        base_salary = 60000
        bonus = self.experience * 5000
        return base_salary + bonus

class Developer(Employee):
    def calculate_salary(self):
        base_salary = 50000
        bonus = self.experience * 4000
        return base_salary + bonus

class Designer(Employee):
    def calculate_salary(self):
        base_salary = 45000
        bonus = self.experience * 3000
        return base_salary + bonus

# Demonstrate polymorphism
employees = [
    Manager("John Smith", "Manager", 5),
    Developer("Alice Johnson", "Developer", 3),
    Designer("Bob Brown", "Designer", 2)
]

for employee in employees:
    print(f"{employee.name} ({employee.role}) - Salary: ${employee.calculate_salary()}")
```

```python
'''18.Discuss the concept of function pointers and how they can be used to achieve polymorphism in Python.'''

# Ans
'''
Objects and Methods:
In Python, functions and methods are objects. Methods are essentially functions bound to objects. This means that you can
pass methods around just like any other object (e.g., variables, lists, etc.).

Dynamic Dispatch:
Dynamic dispatch, often referred to as dynamic method dispatch or late binding, is a mechanism where the method to be invoked
is determined at runtime based on the actual type of the object. This is how polymorphism is achieved in Python.
When you call a method on an object, Python looks up the appropriate method in the object's class and executes it. 
This allows different classes to have methods with the same name but different implementations, enabling polymorphic behavior.

'''

# Code
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

# Polymorphic function
def make_animal_speak(animal):
    return animal.speak()

dog = Dog()
cat = Cat()
print(make_animal_speak(dog)) 
print(make_animal_speak(cat)) 
```

```python
'''20.Create a Python class for a zoo simulation, demonstrating polymorphism with different animal types 
(e.g., mammals, birds, reptiles) and their behavior (e.g., eating, sleeping, making sounds).'''

# Code
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def make_sound(self):
        return "Animal sound"

    def eat(self):
        return "Animal eats"

    def sleep(self):
        return "Animal sleeps"

class Mammal(Animal):
    def make_sound(self):
        return "Mammal sound"

    def eat(self):
        return "Mammal eats"

class Bird(Animal):
    def make_sound(self):
        return "Bird sound"

    def eat(self):
        return "Bird eats"

class Reptile(Animal):
    def make_sound(self):
        return "Reptile sound"

    def eat(self):
        return "Reptile eats"

# Zoo simulation
zoo = [
    Mammal("Lion", "Mammal"),
    Bird("Eagle", "Bird"),
    Reptile("Snake", "Reptile"),
]

for animal in zoo:
    print(f"{animal.name} ({animal.species}):")
    print(f"Sound: {animal.make_sound()}")
    print(f"Eating: {animal.eat()}")
    print(f"Sleeping: {animal.sleep()}")
    print()
```
